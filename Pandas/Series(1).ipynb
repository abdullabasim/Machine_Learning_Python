```python
df = df[df.Stars != 'Unrated']
df.Stars = df.Stars.astype(float) #convert string to float

df.loc[df.Stars > 2.5]
```

```python


df[(df.Country == 'Taiwan') & (df.Stars > 2.5)]
```

```python
df.loc[(df.Country == 'Taiwan') & (df.Stars > 2.5) ,['Brand','Style']] #isin use between 2 values
```
