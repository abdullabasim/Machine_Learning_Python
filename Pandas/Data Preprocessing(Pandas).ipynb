##### Replace Values in Column Based on Condition in Pandas

```python
# data
Student = {
    'Name': ['John', 'Jay', 'sachin', 'Geetha', 'Amutha', 'ganesh'],
    'gender': ['male', 'male', 'male', 'female', 'female', 'male'],
    'math score': [50, 100, 70, 80, 75, 40],
    'test preparation': ['none', 'completed', 'none', 'completed',
                         'completed', 'none'],
}
  
# creating a Dataframe object
dfReplace = pd.DataFrame(Student)
  
# Applying the condition
dfReplace.loc[dfReplace["gender"] == "male", "gender"] = 1

dfReplace.loc[dfReplace["gender"] == "female" ,"gender"] = 0

dfReplace

#replace using Pandas Mask
dfReplaceMask = pd.DataFrame(Student)
dfReplaceMask['gender'].mask(dfReplaceMask['gender'] == 'female', 0, inplace=True)

dfReplaceMask['gender'].mask(dfReplaceMask['gender'] == 'male',5,inplace=True)
dfReplaceMask

#replace using Pandas replace
dfReplaceReplace = pd.DataFrame(Student)
dfReplaceReplace['Name'].replace("Amu","xyz", inplace=True,regex=True) # regex=True is where like in sql

dfReplaceReplace

#replace using Numpy
#dfReplaceNumpy = pd.DataFrame(Student)
#dfReplaceNumpy['gender'] = np.where(dfReplaceNumpy['gender'] == 'female',0,1) # if go to 0 else go to 1
#dfReplaceNumpy




```
