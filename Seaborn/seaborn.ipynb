# Line Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_feb = flights [flights['month'] == 'Feb']
flights_feb
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))
sns.lineplot(x='year',y='passengers',data=flights,hue='month')
```

```python
fmri = sns.load_dataset('fmri')

fmri
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri ,hue='event' )
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri ,hue='event',kind='line',col='region',row='event')
```

## Scatter Plots

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex' ,style='smoker',size='size',sizes=(100,300))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,6))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex' ,style='smoker',size='size',palette='plasma')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time',style="time",markers={"Lunch" : "s","Dinner" :"X"})
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time',style="sex",kind="scatter",col="day")
```

# Regression Plots

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
tips['size'].value_counts()
```

```python
sns.regplot(data=tips,x='size',y='tip',x_jitter=0.05)
```

```python
sns.regplot(data=tips,x='size',y='tip',x_jitter=0.05,x_estimator=np.mean)
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='smoker') #abile to add hue and col ...etc
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='smoker' ,col='smoker',row='sex',scatter_kws={"s":100,"edgecolor":'blue'})
```

```python
tips['smoker'] = tips['smoker'] == "Yes" # change string to boolen value to use in sns
tips
```

```python
sns.lmplot(data=tips,x='total_bill',y='smoker',logistic=True) # use for logistic classification
```

# Distribution Plots - Histograms

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.histplot(data=penguins,x='flipper_length_mm') #histogram chart 
```

```python
sns.histplot(data=penguins,x='flipper_length_mm',bins=10) #histogram chart ... bins number of columns
```

```python
sns.histplot(data=penguins,x='species',shrink=0.8) #histogram chart for charegorical data
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=penguins,x='flipper_length_mm',hue='species')
```

```python
sns.histplot(data=penguins,x='flipper_length_mm',kde=True)
```

```python
sns.histplot(data=penguins,x='flipper_length_mm',hue="species",kde=True,stat="density",common_norm=True) #group the result
```

```python
sns.countplot(penguins)
```

```python
sns.countplot(data = penguins ,x='body_mass_g')
```

```python
sns.set(style='whitegrid')
fmri = sns.load_dataset("fmri")
 
sns.boxplot(x="timepoint",
                y="signal",
            linewidth=2.5,
                data=fmri)
```
